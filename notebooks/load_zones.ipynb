```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# DB credentials 
server = 'YOUR SERVER NAME'
database = 'NYCTaxiAnalytics'
```

```python
engine = create_engine(
    f"mssql+pyodbc://@{server}/{database}?driver=ODBC+Driver+17+for+SQL+Server"
)
```

```python
url = "https://d37ci6vzurychx.cloudfront.net/misc/taxi_zone_lookup.csv"
```

```python
# Load data
df = pd.read_csv(url)
```

```python
df.isnull().sum()
```

```python
before = len(df)
df = df.dropna()
after = len(df)
print(f"Removed {before - after} rows containing nulls")
```

```python
df.to_sql(
    name='taxi_zones',
    con=engine,
    if_exists="replace",
    index=False
)
```
